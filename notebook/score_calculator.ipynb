```python
import pandas as pd

import os
import glob
from collections import defaultdict

few_shots = [
    'ensemble_with_entropy_eurosat.csv', 
'ensemble_with_entropy_Flower102.csv',  
'ensemble_with_entropy_Cars.csv',  
 'ensemble_with_entropy_DTD.csv',  
'ensemble_with_entropy_UCF101.csv',  
'ensemble_with_entropy_Caltech101.csv',  
'ensemble_with_entropy_Aircraft.csv',  
'ensemble_with_entropy_Food101.csv',  
'ensemble_with_entropy_Pets.csv',  
'ensemble_with_entropy_SUN397.csv'
]
root = '/home/seongha/TPT/notebook/ensemble_with_entropy/'
save_root = '/home/seongha/TPT/notebook'

if not os.path.exists(os.path.join(save_root, 'result')):
    os.makedirs(os.path.join(save_root, 'result'), exist_ok=True)

save_format = 'zsh_{}_{}_{}.csv'
for arch in ['RN50', 'ViT-B/16']:
    print(arch)
    for K in [2,4,8,16,32,64]:
        print(K)
        few_score_dict = defaultdict(list)
        ood_score_dict = defaultdict(list)
        for seed in [0,1,2]:
            path = os.path.join(root, arch, str(seed), str(K))
            extension = 'csv'
            try:
                os.chdir(path)
            except:
                continue
            result = glob.glob('*.{}'.format(extension))
            # print(result) # files list
            for dataset in result:
                df = pd.read_csv(os.path.join(path, dataset))
                if str(dataset) in few_shots:
                    dataset = dataset.split('.')[0].split('_')[-1]
                    few_score_dict[dataset].append(float(df.iloc[-1]['accuracy']))
                else:
                    dataset = dataset.split('.')[0].split('_')[-1]
                    ood_score_dict[dataset].append(float(df.iloc[-1]['accuracy']))

        if few_score_dict.values():
            print("     Few shot")
            for k, v in few_score_dict.items():
                # if len(v)<3: print("more exp ", k, len(v))
                few_score_dict[k] = round(100 * sum(v)/len(v) , 2)
            for k, v in few_score_dict.items():
                print( "Data {},  Accuracy {:.2f}".format(k, v))
            few_score_dict['average'] = round(sum(list(few_score_dict.values()))/len(few_score_dict.values()), 2)
            print("     Total Accuarcy: {:.4f}".format(sum(list(few_score_dict.values()))/len(few_score_dict.values())))
            acc = pd.DataFrame.from_dict([few_score_dict])
            acc.to_csv(os.path.join(save_root, 'result', save_format.format(arch.replace('/', ''), K, 'fewshot')))
        if ood_score_dict.values():
            print("    OOD")
            for k, v in ood_score_dict.items(): ood_score_dict[k] = round(100 * sum(v)/len(v) , 2)
            for k, v in ood_score_dict.items():
                print( "Data {},  Accuracy {:.2f}".format(k, v))
            ood_score_dict['average'] = round(sum(list(ood_score_dict.values()))/len(ood_score_dict.values()), 2)
            print("     Total Accuarcy: {:.4f}".format(sum(list(ood_score_dict.values()))/len(ood_score_dict.values())))
            acc = pd.DataFrame.from_dict([ood_score_dict])
            acc.to_csv(os.path.join(save_root, 'result', save_format.format(arch.replace('/', ''), K, 'ood')))
            
```
