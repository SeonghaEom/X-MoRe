```python
import pandas as pd

import os
import glob
from collections import defaultdict

path = '/home/seongha/TPT/notebook/threshold_with_entropy/RN50/0/2'

root = '/home/seongha/TPT/notebook/threshold_with_entropy/'
for arch in ['RN50', 'ViT-B/16']:
    print(arch)
    for K in [2,4,8,16,32,64]:
        print(K)
        score_dict = defaultdict(list)
        for seed in [0,1,2]:
            path = os.path.join(root, arch, str(seed), str(K))
            extension = 'csv'
            try:
                os.chdir(path)
            except:
                continue
            result = glob.glob('*.{}'.format(extension))
            # print(result) # files list
            for dataset in result:
                df = pd.read_csv(os.path.join(path, dataset))
                score_dict[dataset].append(float(df.iloc[-1]['accuracy']))
        if score_dict.values():
            for k, v in score_dict.items(): score_dict[k] = 100 * sum(v)/len(v) 
            for k, v in score_dict.items():
                print( "Data {},  Accuracy {:.2f}".format(k, v))
            print("Total Accuarcy: {:.4f}".format(sum(list(score_dict.values()))/len(score_dict.values())))
                
            
```
